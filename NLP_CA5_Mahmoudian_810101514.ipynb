## handling duplicates

````python
from collections import defaultdict
import re

#helper
def normalize_country_name(name: str):
    name = name.upper()
    name = re.sub(r"\(.*?\)", "", name)   # remove (...)
    name = name.split("|")[0]             # remove after |
    return name.strip()

reverse = defaultdict(list)

for raw, row in zip(currency_df["Country"], currency_df["Currency_Code"]):
    normalized = normalize_country_name(raw)
    reverse[normalized].append(raw)

duplicates = {k: v for k, v in reverse.items() if len(v) > 1}

duplicates
````

````python
country_to_currency = {}

for _, row in currency_df.iterrows():
    country = str(row["Country"])
    currency = str(row["Currency_Code"]).strip().upper()

    country = normalize_country_name(country)

    if country and country != "NAN" and currency and currency != "NAN":
        if country not in country_to_currency:
            country_to_currency[country] = []

        if currency not in country_to_currency[country]:
            country_to_currency[country].append(currency)
````

## title detection used in indexing the book

````python
def norm(s: str):
    return " ".join(s.strip().split())

def is_all_caps_semantic(line: str):
    has_letter = False
    for ch in line:
        if ch.isalpha():
            has_letter = True
            if ch != ch.upper():
                return False
    return has_letter


def detect_title(lines, i, countries, cities):
    line = norm(lines[i])
    if not line:
        return None, i

    if line in countries or line in cities:
        return None, i

    if not is_all_caps_semantic(line):
        return None, i

    if len(line.split()) > 10:
        return None, i
    title_lines = [line]
    j = i + 1

    # allow second line
    if j < len(lines):
        nxt = norm(lines[j])
        if (
            nxt
            and nxt not in countries
            and nxt not in cities
            and is_all_caps_semantic(nxt)
            and len(nxt.split()) <= 10
        ):
            title_lines.append(nxt)
            j += 1

    return " ".join(title_lines), j
````

## detect cities used in indexing the book

````python
def detect_city(lines, i, cities):
    line = norm(lines[i])

    # single line city
    if line in cities:
        return line, i + 1

    # two line city
    if i + 1 < len(lines):
        line2 = norm(lines[i + 1])
        combined = f"{line} {line2}"
        if combined in cities:
            return combined, i + 2

    return None, i
````

## indexing the book

````python
rows = []
buffer = []

current_country = None
current_city = None
current_title = "DESCRIPTION"

started = 0
finished = False

for doc in documents:
    if finished:
        break

    lines = doc.text.split("\n")
    i = 0

    while i < len(lines):
        line = norm(lines[i])

        if not line:
            i += 1
            continue

        # stop condition
        if line == "Acknowledgments" and started == 2:
            finished = True
            break

        # start condition
        if started != 2:
            if line == "Argentina" and started == 1:
                started = 2
                current_country = "Argentina"
                current_city = None
                current_title = "DESCRIPTION"
            elif line == "Argentina" and started == 0:
                started = 1
            i += 1
            continue

        # detecting country
        if line in COUNTRIES:
            if buffer:
                rows.append({
                    "text": " ".join(buffer),
                    "country": current_country,
                    "city": current_city,
                    "title": current_title
                })
                buffer = []

            current_country = line
            current_city = None
            current_title = "DESCRIPTION"
            i += 1
            continue

        # detecting city
        city, next_i = detect_city(lines, i, CITIES)
        if city:
            if buffer:
                rows.append({
                    "text": " ".join(buffer),
                    "country": current_country,
                    "city": current_city,
                    "title": current_title
                })
                buffer = []

            current_city = city
            current_title = "DESCRIPTION"
            i = next_i
            continue

        # detecting title
        title, next_i = detect_title(lines, i, COUNTRIES, CITIES)
        if title:
            if buffer:
                rows.append({
                    "text": " ".join(buffer),
                    "country": current_country,
                    "city": current_city,
                    "title": current_title
                })
                buffer = []

            current_title = title
            i = next_i
            continue

        # other lines
        buffer.append(line)
        i += 1

if buffer and started and not finished:
    rows.append({
        "text": " ".join(buffer),
        "country": current_country,
        "city": current_city,
        "title": current_title
    })

df = pd.DataFrame(rows)
````

````python
from langchain_core.tools import tool
from datetime import datetime
import os, json

import lancedb
import numpy as np
from tavily import TavilyClient

DB_PATH = "/content/drive/MyDrive/NLP_HW5_Q1/lancedb_travel"
BOOK_TABLE_NAME = "book_vectors"


def compute_days(arrival_date: str, departure_date: str):
    a = datetime.strptime(arrival_date, "%Y-%m-%d")
    d = datetime.strptime(departure_date, "%Y-%m-%d")
    return (d - a).days


@tool(
    description=(
        "Trip planning context builder. Uses LanceDB RAG (book_vectors) to gather destination info using multiple queries. "
        "If book context is empty, uses Tavily fallback for general travel info. Always uses Tavily for interest-based queries if interests provided. "
    )
)
def trip_planning(destination: str, arrival_date: str, departure_date: str, interests: list = None):
    payload = {
        "destination": destination,
        "arrival_date": arrival_date,
        "departure_date": departure_date,
        "interests": interests or [],
        "days": None,
        "rag": {
            "book_results": [],
            "errors": []
        },
        "tavily_fallback": {
            "used": False,
            "results": [],
            "errors": []
        },
        "input_errors": []
    }

    # validate inputs
    destination = (destination or "").strip()
    arrival_date = (arrival_date or "").strip()
    departure_date = (departure_date or "").strip()
    interests = interests or []

    if not destination:
        payload["input_errors"].append("Destination is required.")
        return json.dumps(payload, ensure_ascii=False)

    try:
        datetime.strptime(arrival_date, "%Y-%m-%d")
        datetime.strptime(departure_date, "%Y-%m-%d")
    except Exception:
        payload["input_errors"].append("Invalid date format. Expected YYYY-MM-DD.")
        return json.dumps(payload, ensure_ascii=False)

    days = compute_days(arrival_date, departure_date)
    payload["days"] = days
    if days <= 0:
        payload["input_errors"].append("departure_date must be after arrival_date.")
        return json.dumps(payload, ensure_ascii=False)

    # 1-lanceDB RAG
    rag_queries = [
        f"top attractions in {destination}",
        f"best local food and restaurants in {destination}",
        f"things to do in {destination}",
        f"best areas to stay in {destination}",
    ]
    # add interest RAG queries
    for it in interests:
        rag_queries.append(f"best {it} in {destination}")

    payload["rag"]["queries_used"] = rag_queries

    book_context = []
    try:
        db = lancedb.connect(DB_PATH)
        table = db.open_table(BOOK_TABLE_NAME)

        for q in rag_queries:
            try:
                q_emb = embedding_model.encode(q, normalize_embeddings=True).astype(np.float32)
                rows = (
                    table.search(q_emb)
                    .where(f"city == '{destination}'")
                    .limit(2)
                    .to_list()
                )
                for r in rows or []:
                    book_context.append({
                        "query": q,
                        "title": r.get("title"),
                        "text": r.get("text"),
                    })
            except Exception as e:
                payload["rag"]["errors"].append(f"RAG search error for query '{q}': {type(e).__name__}: {e}")

    except Exception as e:
        payload["rag"]["errors"].append(f"LanceDB connect/open error: {type(e).__name__}: {e}")

    payload["rag"]["book_results"] = book_context

    # 2-tavily fallback
    seen_urls = set()
    tavily = TavilyClient(api_key=os.environ.get("TAVILY_API_KEY"))

    queries = []

    # add general queries only if RAG is empty
    if not payload["rag"]["book_results"]:
        payload["tavily_fallback"]["errors"].append("No RAG data found in the book.")
        payload["tavily_fallback"]["used"] = True

        general_queries = [
            f"top attractions in {destination}",
            f"best local food and restaurants in {destination}",
            f"things to do in {destination}",
            f"best areas to stay in {destination}",
        ]

        for q in general_queries:
            queries.append(q)

    # add interest queries
    for it in interests or []:
        q = f"best {it} in {destination}"
        queries.append(q)

    # tavily search
    for q in queries:
        try:
            resp = tavily.search(
                query=q,
                max_results=1,
                search_depth="advanced"
            )

            for r in (resp.get("results", []) or []):
                url = r.get("url")
                if not url or url in seen_urls:
                    continue

                seen_urls.add(url)

                payload["tavily_fallback"]["results"].append({
                    "query": q,
                    "title": r.get("title"),
                    "url": url,
                    "content": r.get("content"),
                })

        except Exception as e:
            payload["tavily_fallback"]["errors"].append(f"{type(e).__name__}: {e}")

    if not payload["tavily_fallback"]["results"]:
        payload["tavily_fallback"]["errors"].append("No tavily search results found Also.")

    return json.dumps(payload, ensure_ascii=False)
````

````python
from langchain_core.messages import SystemMessage
from langgraph.graph import MessagesState

def reasoner_node(state: MessagesState):
    messages = [SystemMessage(content=SYSTEM_PROMPT)] + state["messages"]
    ai_msg = llm_with_tools.invoke(messages)
    return {"messages": [ai_msg]}
````

````python
from langgraph.graph import StateGraph, END
from langgraph.prebuilt import ToolNode, tools_condition

# Tools node
tools_node = ToolNode(TOOLS)

# Build graph
graph_builder = StateGraph(MessagesState)

graph_builder.add_node("reasoner", reasoner_node)
graph_builder.add_node("tools", tools_node)

# Entry
graph_builder.set_entry_point("reasoner")

graph_builder.add_conditional_edges(
    "reasoner",
    tools_condition,
    {
        "tools": "tools",
        END: END,
    },
)

# go back to reasoner
graph_builder.add_edge("tools", "reasoner")

graph = graph_builder.compile()
print("Graph compiled successfully.")
````

````python
from langchain_core.messages import HumanMessage

def _extract_used_tools(messages):
    used = []
    for m in messages:
        tool_calls = getattr(m, "tool_calls", None)
        if tool_calls:
            for tc in tool_calls:
                name = tc.get("name")
                if name:
                    used.append(name)

    seen = set()
    uniq = []
    for t in used:
        if t not in seen:
            seen.add(t)
            uniq.append(t)
    return uniq

def chat_cli():
    print("Travel Agent CLI. Type 'exit' to quit, 'clear' to reset.\n")

    state = {"messages": []}

    while True:
        user_in = input("You> ").strip()
        if not user_in:
            continue

        if user_in.lower() == "exit":
            print("Bye!")
            break

        if user_in.lower() == "clear":
            state = {"messages": []}
            print("Cleared conversation.\n")
            continue

        # add user message
        state["messages"].append(HumanMessage(content=user_in))

        # invoke graph
        out = graph.invoke(state)

        # update state
        state = out

        # show tools used in this turn
        used_tools = _extract_used_tools(out["messages"])
        if used_tools:
            print(f"[tools used] {', '.join(used_tools)}")

        last_text = None
        for m in reversed(out["messages"]):
            content = getattr(m, "content", None)
            if isinstance(content, str) and content.strip():
                last_text = content
                break

        if last_text is None:
            last_text = "(No assistant text output.)"

        print(f"Agent> {last_text}\n")
````
